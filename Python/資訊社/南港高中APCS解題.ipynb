##三角形辨別

本題是要根據輸入的資料，判斷三個線段是否可以組成一個三角形，如果可以，那會是哪一種三角形?

輸入的資料分別為三個數字代表三個線段長

判斷這三個線段是否能夠形成一個三角形，必須符合「任意兩邊之和大於第三邊」

在檢查時三個邊是否能夠形成三角形時，只要把最長的邊當做是第三邊來判斷即可，

因此，若a,b,c為三個線段的邊長，且c為最大值，則

若 a + b <= c，三線段無法構成三角形。

若 a * a + b * b < c * c，三線段構成銳角三角形。

若 a * a + b * b = c * c，三線段構成直角三角形。

若 a * a + b * b > c * c，三線段構成鈍角三角形。

所以我們可以先把輸入的三個數字由小至大排序，得到最大值，再用if來判斷是哪一種情形。

```
#以下四行為不同的測試資料
#line = '7 4 5'
#line = '3 4 5'
#line = '101 100 99'
line = '10 100 10'

#讀入用空白隔開的字串
data = line.split(' ')

#將串列中的字串轉為數字
data = list(map(int, data))

#將串列中的數字由小到大排序
data.sort()

#將串列中的數字依序取出
a, b, c = data

#判斷是否可以形成三角形
if (a+b<=c):
    result = 'No'
#判斷是否可以形成銳角三角形
elif (a*a + b*b < c*c):
    result = 'Obtuse'
#判斷是否可以形成直角三角形
elif (a*a + b*b == c*c):
    result = 'Right'
#判斷是否可以形成鈍角三角形
elif (a*a + b*b > c*c) :
    result = 'Acute'

#依題目要求格式將結果輸出
print (*data)
print (result)
```

##最大和

給定N群數字，每群都恰有M個正整數，將每群的最大值取出然後相加，就可以得到這N群數字的最大和。

輸入格式

第一行有兩個整數分別代表N群數字，每群各有M個正整數。

接下來每一行代表M個正整數，數字間用空白隔開。

測試資料 maxsum1.txt

測試資料 maxsum2.txt

輸出格式

第一行輸出最大值總和

第二行依序輸出各群最大值中可以整除總和的值，如果都不能整除，就輸出-1。

```
# 從檔案讀取輸入的資料
file = open('d:/maxsum1.txt', "r")

# 讀入第一行用空白隔開的字串
line = file.readline()
data=line.split(' ')

#取得 N(群) 各 M(個)正整數
N, M = map(int,data)

max_number = []
#讀取 N群數據資料
for i in range(N):
    #讀入第i群用空白隔開的正整數資料
    line = file.readline()
    data=line.split(' ')
    #將串列中的字串轉為數字
    data = list(map(int, data))
    #將串列中的數字由小到大排序
    data.sort()
    #將第i群最大數值存到串列中
    max_number.append(data[M-1])

#關閉檔案
file.close()

#將每群數字最大值相加並印出結果
max_sum = sum(max_number)
print(max_sum)

# 找出能夠整除的因數
factors = []
for i in range(N):
    if max_sum % max_number[i] == 0:
        factors.append(max_number[i])

# 依照輸出格式印出結果
if len(factors) > 0 :
    print(*factors)
else :
    print(-1)
```

##定時K彈

給定N個人，每隔M個人會爆炸，總共爆炸K次，爆完的下一個人是幸運者。

輸入格式

只有一行包含三個正整數，依序為N, M, K，中間使用空白隔開。

輸出格式

輸出幸運者的號碼，結尾有換行符號。

![替代文字](http://drweb.nksh.tp.edu.tw/static/lesson/apcs/apcs0071.png)


```
#以下二行是不同的測試資料
#line = '5 2 4'
line = '8 3 6'

#讀入用空白隔開的字串
data = line.split(' ')#將串列中的字串轉為數字並存入相對應的變數
N, M, K = map(int, data)

#產生數字1到N的串列
people = [i for i in range(1, N+1)]

#爆炸者在串列的索引位置
index = 0
for i in range(K):
    #從現在位置向後數M個人，找到爆炸者，將它從串列移除
    #因為每一次爆炸串列會少一個人，所以其索引值也要跟著減1
    index = (index + M - 1) % len(people)
    del people[index]
    #若刪除恰好最末元素，刪除完畢後，此時必須將串列索引值移動到最前頭(0)
    if index == len(people):
        index = 0
print(people[index])
```

##邏輯運算子

本題是要根據輸入的資料，判斷可能是經過哪種邏輯運算所產生的結果?

輸入的資料分別為兩個運算元和一個運算結果

就兩個運算元的值各有0,1兩種，所以經排列後有四種可能，經過三種邏輯運算產生的結果如下：

![替代文字](http://drweb.nksh.tp.edu.tw/static/lesson/apcs/apcs0131.png)

就輸入資料而言，除了兩個運算元的值各有0,1兩種，再加上運算結果的值有0,1兩種，所以排列後有八種可能，我們可以由上表來檢視各組輸入資料可能由哪種邏輯運算產生。

![替代文字](http://drweb.nksh.tp.edu.tw/static/lesson/apcs/apcs0132.png)

因此，我們可以把這八組可能寫成一個字典，來窮舉出所有可能的邏輯運算，以進行逆向工程。

```
# -*- coding: utf-8 -*-
import sys

value_matrix = {
        (0, 0, 0): "AND,OR,XOR",
        (0, 0, 1): "IMPOSSIBLE",
        (0, 1, 0): "AND",
        (0, 1, 1): "OR,XOR",
        (1, 0, 0): "AND",
        (1, 0, 1): "OR,XOR",
        (1, 1, 0): "XOR",
        (1, 1, 1): "AND,OR"
}

for line in sys.stdin:
        t1, t2, t3  = line.split(' ')
        a = int(bool(int(t1)))
        b = int(bool(int(t2)))
        result = int(t3)
        print ('\n'.join(value_matrix[(a, b, result)].split(',')))
```

由於輸入資料可能是大於1的正整數，因此我們將輸入的字串先用int()轉整數後，再使用bool()轉為布林值，最後再用int()將布林值轉整數。
